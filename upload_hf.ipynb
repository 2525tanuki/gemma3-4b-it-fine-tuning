```python
from transformers import AutoModelForImageTextToText, AutoProcessor
import os
```

```python
# ローカルに保存されたモデルとプロセッサのディレクトリ
save_dir = "/content/drive/MyDrive/gemma3-jicvqa-finetuned"

# モデルとプロセッサをロード
model = AutoModelForImageTextToText.from_pretrained(save_dir)
processor = AutoProcessor.from_pretrained(save_dir)
```

```python
from google.colab import userdata
os.environ["sft_from_colab"] = userdata.get("sft_from_colab")

from huggingface_hub import login
login(token=os.environ["sft_from_colab"])
```

```python

# Hugging Face Hub 上のリポジトリ名（例：ご自身のユーザー名とリポジトリ名に変更してください）
repo_name = "2525tanuki/gemma3-jicvqa-sft"

# モデルを HF Hub にアップロード
model.push_to_hub(
    repo_name,
    commit_message="Upload SFT model to HF Hub",
    use_auth_token=True  # 環境変数やログイン済みの状態なら True でOK
)

# プロセッサも同じリポジトリにアップロード（README や tokenizer/config も含む場合）
processor.push_to_hub(
    repo_name,
    commit_message="Upload SFT model processor to HF Hub",
    use_auth_token=True
)

print(f"Model and processor have been successfully uploaded to https://huggingface.co/{repo_name}")
```
